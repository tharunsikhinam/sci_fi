```python
# sub_genre score function
# input: dataframe, subgenre column name, list of keywords
# output: assign score to given sub_genre column:
# alien, robot, dystopian, fantasy, monster, time_travel, techno, super, space
def search(df, sub_genre, keywords):
    """
    Return the dataframe sub_genre column populated with total count of occurence of keywords
    """
    # df[['plot']] = df[['plot']].astype(str)
    for index, row in df.iterrows():
        for word in keywords:
            word_count = pd.Series(df['plot'][index]).str.count(word).sum()
            df[sub_genre][index] = df[sub_genre][index] + word_count
```

```python
col_list= list(df_plots)
col_list.remove('Movie_title')
col_list.remove('imdb_id')
col_list.remove('plot')
df_plots['total'] = df_plots[col_list].sum(axis=1)
```

```python
ratio = 100
df_plots['alien'] = df_plots['alien'] / df_plots['total'] * ratio
df_plots['robot'] = df_plots['robot'] / df_plots['total'] * ratio
df_plots['dystopian'] = df_plots['dystopian'] / df_plots['total'] * ratio
df_plots['fantasy'] = df_plots['fantasy'] / df_plots['total'] * ratio
df_plots['monster'] = df_plots['monster'] / df_plots['total'] * ratio
df_plots['time_travel'] = df_plots['time_travel'] / df_plots['total'] * ratio
df_plots['techno'] = df_plots['techno'] / df_plots['total'] * ratio
df_plots['super'] = df_plots['super'] / df_plots['total'] * ratio
df_plots['space'] = df_plots['space'] / df_plots['total'] * ratio
```

```python
df_plots = df_plots.round({'alien': 2, 'space': 2, 'robot': 2, 'time_travel': 2, 'super': 2,\
                           'dystopian': 2, 'monster': 2, 'techno': 2, 'fantasy': 2})
```
