```python
'''
create new dataframe: 
year, 
total_no_of_movies,
genre1_count
genre2_count
...
genre9_count
'''
df_movies_metadata = df_movies_metadata[['year','sub_genres']]
```

```python
df_movies_metadata.head(10)
# df_movies_metadata = df_movies_metadata.join(df_movies_metadata['sub_genres'].str.split('/', expand=True).add_prefix('sub_genres'))
```

```python
# alien, robot, dystopian, fantasy, monster, time_travel, techno, 
df_movies_metadata['alien'] = np.where(df_movies_metadata['sub_genres'].str.contains('alien'), 1, 0)
df_movies_metadata['robot'] = np.where(df_movies_metadata['sub_genres'].str.contains('robot'), 1, 0)
df_movies_metadata['dystopian'] = np.where(df_movies_metadata['sub_genres'].str.contains('dystopian'), 1, 0)
df_movies_metadata['fantasy'] = np.where(df_movies_metadata['sub_genres'].str.contains('fantasy'), 1, 0)
df_movies_metadata['monster'] = np.where(df_movies_metadata['sub_genres'].str.contains('monster'), 1, 0)
df_movies_metadata['time_travel'] = np.where(df_movies_metadata['sub_genres'].str.contains('time_travel'), 1, 0)
df_movies_metadata['techno'] = np.where(df_movies_metadata['sub_genres'].str.contains('techno'), 1, 0)
df_movies_metadata['super'] = np.where(df_movies_metadata['sub_genres'].str.contains('super'), 1, 0)
df_movies_metadata['space'] = np.where(df_movies_metadata['sub_genres'].str.contains('space'), 1, 0)
```

```python
# Set everything in sub_genres to 1, set column name to movie_count
df_movies_metadata['sub_genres'] = 1
df_movies_metadata.rename(columns={'sub_genres':'movies_count'}, inplace=True)
```

```python
df_movies_metadata.loc[df_movies_metadata['year'].values=='1896']
```

```python
df_movies_agg = df_movies_metadata.groupby('year').sum()

df_movies_agg

# df.rename({'sub_genres':'movie_count', '$b':'b', '$c':'c', '$d':'d', '$e':'e'}, axis=1)
```

```python
ratio = 100
df_movies_agg['alien'] = df_movies_agg['alien'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['robot'] = df_movies_agg['robot'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['dystopian'] = df_movies_agg['dystopian'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['fantasy'] = df_movies_agg['fantasy'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['monster'] = df_movies_agg['monster'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['time_travel'] = df_movies_agg['time_travel'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['techno'] = df_movies_agg['techno'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['super'] = df_movies_agg['super'] / df_movies_agg['movies_count'] * ratio
df_movies_agg['space'] = df_movies_agg['space'] / df_movies_agg['movies_count'] * ratio
```

```python
df_movies_agg = df_movies_agg.round(2)
```

```python
df_movies_agg.to_csv('movies_subgenres_count_by_year_12-1-2018.csv', sep=',')
```
